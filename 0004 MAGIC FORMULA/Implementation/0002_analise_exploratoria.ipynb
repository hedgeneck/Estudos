```python
# O objetivo aqui vai ser consolidar total assets ativas e consolidadas e tirar algumas conclus√µes sobre os tickers
```

```python
import pandas as pd
import numpy as np
```

```python
ta_ativas = pd.read_excel('../../data/economatica_1/total_assets_mensal_ativas.xlsx',
                            skiprows = [1,2],
                            header=1,
                            index_col = 0,
                            na_values='-')
ta_canceladas = pd.read_excel('../../data/economatica_1/total_assets_mensal_canceladas.xlsx',
                            skiprows = [1,2],
                            header=1,
                            index_col = 0,
                            na_values='-')
# tudo sobre merging de dataframes
# https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
```

```python
# retorna um numpy array
#names_ativas.columns.str.extract('(.*)$')[0].values
# retorna uma list
#names_ativas.columns.str.extract('(.*)$')[0].tolist()

ticker_ativas = ta_ativas.columns.str.extract('(.*)$')[0].tolist()
ticker_canceladas = ta_canceladas.columns.str.extract('(.*)$')[0].tolist()

tickers = ticker_ativas + ticker_canceladas
```

```python
len(tickers)
```

```python
len(set(tickers))
```

```python
# para aplicar regular expressions with extract, √© mais f√°cil com series ao inv√©s de lists
ticker_ativas = ta_ativas.columns.str.extract('(.*)$')[0]
ticker_canceladas = ta_canceladas.columns.str.extract('(.*)$')[0]

# absolutamente n√£o √© assim que se concatena duas s√©ries
# tickers = ticker_ativas + ticker_canceladas

tickers = pd.concat([ticker_ativas, ticker_canceladas], ignore_index=True)
```

```python
# https://www.debuggex.com/cheatsheet/regex/python
# https://www.dataquest.io/blog/regex-cheatsheet/
balcao = tickers[tickers.str.contains('.B$')]
units = tickers[tickers.str.contains('.11$')]
on = tickers[tickers.str.contains('.\D3$')]
pn = tickers[tickers.str.contains('.\D4$')]
#pn2 = tickers[tickers.str.contains('.(?:\D3|\D4)$')] # termina apenas em 3 (sem 13) ou 4 (sem 14)
pn2 = tickers[tickers.str.contains('.(?:\D5|\D6|\D7|\D8)$')]
len(pn2)
pn2
```

```python
print("N√∫mero total de tickers: ", len(tickers))
print("N√∫mero total de ON: ", len(on))
print("N√∫mero total de PN 4: ", len(pn))
print("N√∫mero total de outras PN 5 6 7 8: ", len(pn2))
print("N√∫mero total de ON+PN: ", len(on)+len(pn)+len(pn2))
```

```python
stocks = pd.concat([on, pn])
len(stocks)
```
