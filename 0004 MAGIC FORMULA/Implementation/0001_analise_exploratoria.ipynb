```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# carregar o excel tinha dado erro com xlrd library
# um pip install xlrd resolveu
# poderia dar problema de estar instalado pra python 2 e não pra python 3
# se resolveria com python3 -m pip install xlrd
# o pip3 install xlrd seria outra opção mas não sei pq não deu
# https://stackoverflow.com/questions/45244238/importerror-no-module-named-xlrd
# aqui tem uma explicação legal sobre como modificar nome de colunas usando import re
# https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html
# e https://jakevdp.github.io/PythonDataScienceHandbook/03.10-working-with-strings.html
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html
# skip rows https://stackoverflow.com/questions/49801060/skipping-range-of-rows-after-header-through-pandas-read-excel
# index_col = 0
total_assets = pd.read_excel('../../data/economatica_1/total_assets_mensal_ativas.xlsx',
                            skiprows = [1,2],
                            header=1,
                            index_col = 0,
                            na_values='-')
```

```python
# Para mudar o nome das colunas, é preciso um pouco de regular expressions
# https://medium.com/@alegeorgelustosa/an%C3%A1lise-explorat%C3%B3ria-e-preditiva-do-dataset-titanic-em-python-fbd5e5bb7328
# http://mundoia.com.br/tutorial/analise-exploratoria-de-dados-extraindo-insights-do-fifa-18/
# https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html
# https://jakevdp.github.io/PythonDataScienceHandbook/03.10-working-with-strings.html
# https://www.w3schools.com/python/python_regex.asp
# https://docs.python.org/3/library/re.html
# https://stackoverflow.com/questions/1327369/extract-part-of-a-regex-match
```

```python
total_assets = total_assets[total_assets.index.str.contains("Mar|Jun|Sep|Dec")]
# total_assets.columns.str.slice() tira espaços do começo e do fim
# essa linha deu trabalho pra caralho.
# total_assets.columns.str.extract('(.*)$') seria um dataframe com uma coluna e os nomes das columns ficariam
# como (QVQP3B,)	(ABCB4,)	(EALT3,)	(EALT4,)	(ADHM3,)	(TIET3,)
# total_assets.columns.str.extract('(.*)$').values seria como numpy array e ficaria
# como (QVQP3B,)	(ABCB4,)	(EALT3,)	(EALT4,)	(ADHM3,)	(TIET3,)
# total_assets.columns.str.extract('(.*)$')[0] seria como pd series e dá certo
total_assets.columns = total_assets.columns.str.extract('(.*)$')[0]
total_assets.columns.name = "ticker"
total_assets = total_assets.T
```

```python
# algumas contagens interessantes
# número de tickers
print("Número de tickers: ", total_assets.index.size)
print("Numero de trimestres: ", total_assets.columns.size)
# total_assets["Jun-09"] coluna referencia com []. Retorna Series
# total_assets.loc["VVAR3"] para referenciar index usa loc e iloc com []. Retorna pd.Series
```

```python
print("non missing values Mar-87", total_assets["Mar-87"].count())
print("non missing values Jun-09", total_assets["Jun-09"].count())
print("non missing values Dec-17", total_assets["Dec-17"].count())
```

```python
count_series = pd.Series()
count_series.index.name = "Trimestre"
# o nome da series deveria aparecer na y label mas não aparece
# https://github.com/pandas-dev/pandas/issues/18124
# count_series.name = "Non NA count"
for i in total_assets:
    series_ = pd.Series(total_assets[i].count(), index=[i])
    count_series = pd.concat([count_series,series_])
count_series
```

```python
ax = count_series.plot('bar')
# código tirado de https://stackoverflow.com/questions/50675920/tick-labels-overlap-in-pandas-bar-chart
ax.set_xticklabels([t if not i%5 else "" for i,t in enumerate(ax.get_xticklabels())])
ax.plot()
```
